```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/kaggle/input/climate-change-indicators/climate_change_indicators.csv')
# Data Cleaning
df = df.dropna()  # Remove missing values

# Data Exploration
print(df.describe())

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(df['Country'], df['Source'], label='Indicator')
plt.xlabel('Year')
plt.ylabel('Temperature Anomaly (°C)')
plt.title('Global Surface Temperature Anomalies')
plt.legend()
plt.grid(True)
plt.show()

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import squarify

# Sample dataset
data = {
    'Country': ['Afghanistan', 'Albania', 'Algeria', 'Anguilla', 'Bahamas'],
    'Temperature_Anomaly': [0.5, 0.8, 0.3, 0.6, 0.9]
}

df = pd.DataFrame(data)

# Histogram
plt.figure(figsize=(12, 6))
sns.histplot(df['Temperature_Anomaly'], bins=10, kde=True)
plt.xlabel('Temperature Anomaly (°C)')
plt.ylabel('Frequency')
plt.title('Distribution of Temperature Anomalies')
plt.grid(True)
plt.show()

# Bar Chart
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Temperature_Anomaly', data=df)
plt.xlabel('Country')
plt.ylabel('Average Temperature Anomaly (°C)')
plt.title('Average Temperature Anomaly by Country')
plt.grid(True)
plt.show()

# Treemap
plt.figure(figsize=(12, 6))
squarify.plot(sizes=df['Temperature_Anomaly'], label=df['Country'], alpha=0.8)
plt.axis('off')
plt.title('Total Temperature Anomaly by Country')
plt.show()
```
